```python
# To use pandas for the Clean and transform the raw dataset
import pandas as pd

#To Read the csv file
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
# To check the the missing values Present in the columns

df.isnull().sum()
```

```python
# To remove the null values from the columns..
# i.e Review Rating(Column having 37 null value)

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
# To check the the missing values Present in the columns

df.isnull().sum()
```

```python
#To convert the column names into lower case.
df.columns= df.columns.str.lower()

#To replace the space between the column names words into the underscore.
df.columns = df.columns.str.replace(' ','_')

#to replace the name of the column to new name 
# i.e ( purchase_amount_(usd)':'purchase_amount' )
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#To check the columns are converted into the snake case or not ....
# snake case means  (ex 1. customer_id )

df.columns
```

```python
# To create a new column age_group which is depends or based on the age column 

labels =['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
#To check the top 10 rows from the new columns age_group
df[['age','age_group']].head(10)
```

```python
#To create a column purchace_frequency_days based/depend on the frequency_of_purchases


frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#To check the top 10 rows from the new columns purchase_frequency_days
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
#To check the dependancy between the discount_applied and promo_code_used 
# to want the discount then promocode will be used thats why teh columns records are same 

df[['discount_applied','promo_code_used']].head(10)
```

```python
#To check the discount_applied and promo_code_used  ll 
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#To remove the column of promo_code_used because of the both  discount_applied and promo_code_used are the same columns

df = df.drop('promo_code_used', axis=1)
```
